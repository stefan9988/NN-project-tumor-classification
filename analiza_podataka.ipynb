```python
def extract_numbers_between_brackets(string):
    result = []
    start = -1
    end = -1
    flag = False
    for i, char in enumerate(string):
        if char == '(':
            start = i
            flag = True
        elif char == ')':
            end = i
            flag = False
            numbers = [int(x) for x in string[start + 1:end].split(',')]
            result.extend(numbers)
    return result
```

```python
#Izvlacenje brojeva i stavljanje u listu
dim=[]
for i in range(df.shape[0]):
    q=extract_numbers_between_brackets(df["shape"][i])
    dim.append(q)
```

```python
m_new_0=0
m_new_1=0
for i in range(df.shape[0]):
    m_new_0 = m_new_0 + (dim[i][1] - m_new_0) / (i + 1)
    m_new_1 = m_new_1 + (dim[i][0] - m_new_1) / (i + 1)
```
